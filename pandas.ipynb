# pd.isnull

```python
data = {'a': [1,3,5,None],
        'b': [7,9,None,11],
        'c': [13,None,15,17]}

df=pd.DataFrame(data)

null_values = pd.isnull(df)

print(null_values)
```

# `pd.notnull

```python
data = {'A': [1,3,5,None],
        'B': [7,9,None,11],
        'C': [13,None,15,17]}

df = pd.DataFrame(data)

not_null_values = pd.notnull(df)

print(not_null_values)
```

# df.dropna()

```python
data = {'A': [1,3,5,None],
        'B': [7,9,None,11],
        'C': [13,None,15,17]}

df = pd.DataFrame(data)

clean_df = df.dropna()

print(clean_df)
```

# df.dropna(axis=1)

```python
data = {'A': [1,3,5,None],
        'B': [7,9,None,11],
        'C': [13,6,15,17]}

df = pd.DataFrame(data)

clean_df = df.dropna(axis=1)

print(clean_df)
```

# df.dropna(axis=1, thresh=n)

```python
data = {'A': [1,3,5,None],
        'B': [None,9,None,11],
        'C': [13,6,15,17]}

df = pd.DataFrame(data)

clean_df = df.dropna(axis=1,thresh=3)

print(clean_df)
```

```python
data = {'A': [3,5,None,None],
        'B': [9,None,11,12],
        'C': [None,15,17,None]}

df = pd.DataFrame(data)

clean_df = df.dropna(axis=1,thresh=3)

print(clean_df)
```

```python
data = {'A': [3,5,None,None],
        'B': [9,None,11,12],
        'C': [None,15,17,None]}

df = pd.DataFrame(data)

clean_df = df.dropna(axis=1,thresh=2)

print(clean_df)
```
